## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

```python
#checking for percentage of missing data since there are a lot of columns
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
#Check for null values in a different way
df.isnull().sum()
```

```python
#Drop rows with null because I don't want to have so many nulls in my analysis
modifiedData = df.dropna(axis = 0, how ='any') 
modifiedData 
```

```python
#Check for dataframe comparison
#A lot of rows were eliminated. I wonder if this makes my data not very reliable. 

print("Old data frame length:", len(df))
print("New data frame length:", len(modifiedData)) 
print("Number of rows with at least 1 NA value: ",
      (len(df)-len(modifiedData)))
```

```python
cols = df.columns[:50] #Checking for missing values.  Not many are showing because Ive gotten rid of the null values above.
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```
